```python
# preprocess merged_jobs.csv, company-wikiurl.json
# input file: merged_jobs.csv
# output: cleaned_jobs.csv, companyRDF.csv
import pandas as pd
df = pd.read_csv("merged_jobs.csv")
# company name: make_lower case
companyName = [com.lower() for com in df["Company Name"]]
df["Company Name"] = companyName
df.sort_values(by= 'Company Name', axis=0, ascending=True, inplace=True, kind='quicksort', na_position='last')
df.reset_index(drop=True, inplace=True)
```

```python
import json
with open('company-wikiurl.json') as f:
    data = json.load(f)
```

```python
wikisite = list()
companyName = df["Company Name"] 

for com in companyName:
    if com in data:
        wikisite.append(data[com])
    else:
        wikisite.append(None)
```

```python
df['wikisite'] = wikisite
```

```python
import re

def uri_fuc(key):
    newkey = re.sub(r" corporation", "", key)
    newkey = re.sub(r" incorporation", "", newkey)
    newkey = re.sub(r"[.,<>&()]", "",newkey)
    newkey = re.sub(r"[/-]", " ",newkey)
    newkey = re.sub(r" and ", " ", newkey)
    newkey = re.sub(r"  ", " ", newkey)
    newkey = re.sub(r"technologies", "technology", newkey)
    newkey = re.sub(r"laboratories", "labs", newkey)  
    # newkey = re.sub(r" ", "_",newkey)
    words = re.split(" ", newkey)[:3]
    new_name = ""
    for w in words:
        new_name += w+"_"
    return new_name[:-1]
```

```python
companyList = df['Company Name']
name = list()
for com in companyList:
    #node_uri = URIRef(MYNS[uri_fuc(com)])
    name.append(uri_fuc(com))
```

```python
df['uri_name'] = name
```
