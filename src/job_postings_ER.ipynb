- Job title should match

```python
# USE JARO WINKLER
from rltk.similarity.jaccard import jaccard_index_similarity
from rltk.similarity.levenshtein import levenshtein_similarity
from rltk.similarity.jaro import jaro_winkler_similarity
import nltk
jaro_winkler_similarity('boys town', 'boys town hospital')
jaro_winkler_similarity('act consulting', 'CyberCoders')
```

```python
def company_name_sim(c1, c2):
    if pd.isna(c1) or pd.isna(c2):
        return 0
    return jaro_winkler_similarity(c1, c2)

def company_url_sim(u1, u2):
    return 1 if u1 == u2 else 0

def job_title_sim(t1, t2):
    if pd.isna(t1) or pd.isna(t2):
        return 0
    return jaro_winkler_similarity(t1, t2)

def job_description_sim(d1, d2):
    if pd.isna(d1) or pd.isna(d2):
        return 0
    return rltk.ngram_similarity(d1, d2)

def location_sim(l1, l2):
    if pd.isna(l1) or pd.isna(l2):
        return 0
    s1, s2 = set(nltk.word_tokenize(l1)), set(nltk.word_tokenize(l2))
    return jaccard_index_similarity(s1, s2)
```

```python
location_sim("Los Angeles, CA", "Neceville, FL")
job_title_sim("Senior Software Engineer", "Data Scientist")
```

```python
# score in a naive even-split way
er_results = []
for index, row in monster_df.iterrows():
    company_name, location, job_title = row['Company Name'], row['Location'], row['Job Title']
    for idx, g_row in glassdoor_df.iterrows():
        if row['SE/DS'] != g_row['SE/DS']:
            continue
        g_company_name, g_location, g_job_title = g_row['Company Name'], g_row['Location'], g_row['Job Title']
        location_score = location_sim(location, g_location)
        if location_score < 0.8:
            continue
        else:
            company_name_score = company_name_sim(company_name, g_company_name)
            if company_name_score < 0.75:
                continue
            job_title_score = job_title_sim(job_title, g_job_title)
            score = (company_name_score + job_title_score + location_score)/3
            er_results.append((score, company_name_score, job_title_score, index, idx))
```

```python
er_results.sort(key=lambda x: x[0], reverse=True)
```

```python
# When multiple entries are resolved: take the one with the highest score
er_results_dict = {}
for score, company_name_score, job_title_score, index, idx in er_results:
    if index not in er_results_dict:
        er_results_dict[index] = idx
    else:
        continue
er_results_dict
```

```python
# Read in company URL
company_urls = pd.read_csv(f"{HOME_DIR}/data/companyURL.csv")
company_urls = company_urls.drop_duplicates(subset=["Company Name"])
company_urls
```

```python
# join company urls with glassdoor data
merged_glassdoor_df = pd.merge(glassdoor_df, company_urls, left_on="Company Name", right_on="Company Name", how="left")
merged_glassdoor_df['Website'] = merged_glassdoor_df['Website'].fillna(-1)
merged_glassdoor_df["Company Description"] = ""
merged_glassdoor_df
```

```python
start_index = len(merged_glassdoor_df.index)
result_df = merged_glassdoor_df
for i in range(len(monster_df.index)):
    m_row = monster_df.iloc[i]
    if i in er_results_dict:
        print(f"Updating row {i} as {er_results_dict[i]}...")
        if not pd.isna(m_row['Company Description']):
            result_df.loc[er_results_dict[i], 'Company Description'] = m_row['Company Description']
        if not pd.isna(m_row['Website']) and merged_glassdoor_df.loc[er_results_dict[i], 'Website'] == -1:
            result_df.loc[er_results_dict[i], 'Website'] = m_row['Website']
    else:
        # print(f"Inserting row {i}...")
        new_row = {
            "Industry": m_row['Industry'],
            "Job Description": m_row['Job Description'],
            "Job Title": m_row['Job Title'],
            "Location": m_row['Location'],
            "Rating": -1,
            "Revenue": -1,
            "Salary Estimate": -1,
            "Sector": -1,
            "Size": m_row['Size'],
            "Type of ownership": -1,
            "SE/DS": m_row['SE/DS'],
            "Company Name": m_row['Company Name'],
            "Competitors": -1,
            "Founded": m_row['Founded'],
            "Website": m_row['Website'],
            "Headquarters": m_row['Headquarters'],
            "Company Description": m_row['Company Description']
        }
        result_df = result_df.append(new_row, ignore_index=True)
result_df.fillna(-1)
result_df
```
