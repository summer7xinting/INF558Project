```python
def get_jobs(driver, verbose):
    
    '''Gathers jobs as a dataframe, scraped from Glassdoor'''
# Take the current driver state: manual scroll down
#     url = f'https://www.monster.com/jobs/search/?q=Software-Engineer&stpage={num_pages}&page={num_pages}'
#     driver.get(url)
    jobs = []


# Let the page load. Change this number based on your internet speed.
# Or, wait until the webpage is loaded, instead of hardcoding it.
#         time.sleep(4)

        #Test for the "Sign Up" prompt and get rid of it.
#         try:
#             driver.find_element_by_class_name("selected").click()
#         except ElementClickInterceptedException:
#             pass

#         time.sleep(.1)

#         try:
#             driver.find_element_by_class_name("ModalStyle__xBtn___29PT9").click()  #clicking to the X.
#         except NoSuchElementException:
#             pass

        
    #Going through each job in this page
    job_buttons = driver.find_elements_by_xpath("//section[@data-postingid]")  #jl for Job Listing. These are the buttons we're going to click.
    for job_button in job_buttons:  
        time.sleep(1)
        print("Progress: {}".format("Completed " + str(len(jobs)) + " jobs!"))

        job_button.click()  #You might 
        time.sleep(1)
        counter = 0

        while counter < 10:
            try:
                location = driver.find_element_by_xpath("//div[@class='heading']/h2[@class='subtitle']").text
                job_title = driver.find_element_by_xpath("//div[@class='heading']/h1[@class='title']").text
                job_description = driver.find_element_by_xpath("//div[@id='JobDescription']").text
                job_id = driver.current_url.split("&jobid=")[-1]
                counter = 11
            except:
                time.sleep(5)
                counter += 5
        if counter == 10:
            continue

#             try:
#                 salary_estimate = driver.find_element_by_xpath('.//span[@class="gray small salary"]').text
#             except NoSuchElementException:
#                 salary_estimate = -1 #You need to set a "not found value. It's important."

#             try:
#                 rating = driver.find_element_by_xpath('.//span[@class="rating"]').text
#             except NoSuchElementException:
#                 rating = -1 #You need to set a "not found value. It's important."

        #Printing for debugging
        if verbose:
            print("Job Title: {}".format(job_title))
            print("Job Description: {}".format(job_description[:500]))
            print("Location: {}".format(location))

        #Going to the Company tab...
        company_name = None
        company_description = None
        headquarters = None
        size = None
        founded = None
        industry = None
        website = None

        try:
            driver.find_element_by_xpath("//a[@href='#tab-company']").click()
            time.sleep(4)
            company_name = driver.find_element_by_xpath("//div[@id='AboutCompany']/header/h2").text
            company_description = driver.find_element_by_xpath("//div[@id='AboutCompany']//div[@id='AboutCompanyDescription']").text

            keys = driver.find_elements_by_xpath("//div[@id='AboutCompany']//dl/dt")
            values = driver.find_elements_by_xpath("//div[@id='AboutCompany']//dl/dd")

            assert len(keys) == len(values), f"Something is wrong with the tabular format!"

            for idx in range(len(keys)):
                key, value = keys[idx], values[idx]
                if key.text == "Website":
                    website = value.text
                elif key.text == "Founded":
                    founded = value.text
                elif key.text == "Company Size":
                    size = value.text
                elif key.text == "Industry":
                    industry = value.text
                elif key.text == "Headquarters":
                    headquarters = value.text
                else:
                    continue

        except:  #Rarely, some job postings do not have the "Company" tab.
            pass

        if verbose:
            print("Headquarters: {}".format(headquarters))
            print("Size: {}".format(size))
            print("Founded: {}".format(founded))
            print("Industry: {}".format(industry))
            print("Website: {}".format(website))
            print("@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@@")

        jobs.append({
            "Job Title" : job_title,
            "Job Description" : job_description,
            "Job Id" : job_id,
            "Location" : location,
#                 "Salary Estimate" : salary_estimate,
            "Company Name" : company_name,
            "Company Description": company_description,
            "Headquarters" : headquarters,
            "Size" : size,
            "Founded" : founded,
            "Industry" : industry,
            "Website" : website,
        })

        #Clicking on the "next page" button
#         try:
#             driver.find_element_by_xpath('.//li[@class="next"]//a').click()
#         except NoSuchElementException:
#             print("Scraping terminated before reaching target number of jobs. Needed {}, got {}.".format(num_jobs, len(jobs)))
#             break

    return pd.DataFrame(jobs)  #This line converts the dictionary object into a pandas DataFrame.
```

### Clean up csv files

```python
import pandas as pd
df = pd.read_csv("monster_sde.csv")
df
```

```python
# for each row: if company_name is None: split job title and fill in
for index, row in df.iterrows():
    job_title = row['Job Title'] 
    if " at " in job_title:
        delimiter = f" at "
    elif " from " in job_title:
        delimiter = f" from "
    else:
        continue
    # print(f"Splitting {job_title} and {company_name} with delimiter {delimiter}")
    sentence_frags = job_title.split(delimiter)
    new_title, new_company = sentence_frags[0], sentence_frags[-1]
    # print(f"new title is {new_title}, new company is {new_company}")
    if pd.isna(row['Company Name']):
        df.at[index, 'Company Name'] = new_company
    df.at[index, 'Job Title'] = new_title
df
```
